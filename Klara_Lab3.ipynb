```python
#import the packages needed
import rasterio
from rasterio import plot
import matplotlib.pyplot as plt
from rasterio.merge import merge
```

```python
# Use Rasterio to open the image
image1 = rasterio.open("image1.tif")
image2 = rasterio.open("image2.tif")
```

```python
#Lab 3 Part 1: Plot(s) that show both of the Sentinel-2 scenes using the terrain colormap.
#plot the image using the 'terrain' color map
plot.show(image1, cmap='terrain')
```

```python
plot.show(image2, cmap='terrain')
```

```python
#Code that confirms that both Sentinel-2 scenes have the same coordinate reference system. 
#Use a Markup cell to show the name of this CRS.
image1.crs
image2.crs
#Hint: They are the same
#WGS 84 / UTM Zone 18N
```

```python
#plot each band of the image individually
fig, (axr, axg, axb, axn) = plt.subplots(1,4, figsize=(20,5))
plot.show((image1, 3), ax=axr, cmap='Reds', title='red channel')
```

```python
#Some Quick notes
#https://gis.stackexchange.com/questions/247231/rasterio-tool-for-creating-mosaic/247430
#3rd Down for reference
#
#Code that creates a mosaic image from image1 and image2 and plots it using the terrain colormap.
tif_to_mosaic = [image1, image2] #creates array of images to mosaic
```
